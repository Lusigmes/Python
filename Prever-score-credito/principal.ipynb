## SCORE DE CRÉDITO
#### Definir automaticamente após uma análise de dados se o score de crédito do cliente é : Bom, Ok, Ruim.

```python
import pandas as pd
tabela = pd.read_csv("clientes.csv")
tabela = tabela.drop("id_cliente",axis=1)
display(tabela)
```

```python
#Verificar informações sobre a tabela
display(tabela.info()) # tabela = tabela.dropna()
display("------------------------------------------------")
display(tabela.columns)
```

```python
#Preparar a base para IA com labelEncoder
from sklearn.preprocessing import LabelEncoder

coder = LabelEncoder()
for coluna in tabela.columns:
    if tabela[coluna].dtype ==  "object" and coluna != "score_credito":
        tabela[coluna] = coder.fit_transform(tabela[coluna])
        display(coluna)
display("------------------------------------------------")
# display(tabela["profissao"].value_counts())
display(tabela.info())
```

```python
#Machine Learning
from sklearn.model_selection import train_test_split
#treinar e testr a ia
y = tabela["score_credito"] # y - coluna para prever
x = tabela.drop(columns=["score_credito"]) # x - coluna usada para fazer a previsao
# display(tabela)

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y)
```

```python
# Árvore de decisão e Vizinhos proximos
#importa -> criar -> treinar IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

arvore_decisao = RandomForestClassifier()
vizinhos_proximos = KNeighborsClassifier()

arvore_decisao.fit(x_treino,y_treino)
vizinhos_proximos.fit(x_treino,y_treino)
```

```python
count_scores = tabela["score_credito"].value_counts()
display(count_scores)
display("------------------------------------------------")
display(count_scores["Standard"]/sum(count_scores)) 
```

```python
from sklearn.metrics import accuracy_score
#testar as previsoes
previsao_arvore_decisao = arvore_decisao.predict(x_teste)
previsao_vizinhos_proximos = vizinhos_proximos.predict(x_teste.to_numpy())

display(accuracy_score(y_teste,previsao_arvore_decisao))
display(accuracy_score(y_teste,previsao_vizinhos_proximos))
```

```python
nova_tabela = pd.read_csv("novos_clientes.csv")

for coluna in nova_tabela.columns:
    if nova_tabela[coluna].dtype ==  "object" and coluna != "score_credito":
        nova_tabela[coluna] = coder.fit_transform(nova_tabela[coluna])

display(nova_tabela)

display("---------------------arvore decisão---------------------------")
previsao_ad = arvore_decisao.predict(nova_tabela)
display(previsao_ad)

display("---------------------vizinhos proximos------------------------")
previsao_vp =  vizinhos_proximos.predict(nova_tabela)
display(previsao_vp)
```

```python
#Colunas que possuem mais relevancia na hora de definir o score de credito
colunas = list(x_teste.columns)
colunas_importantes = pd.DataFrame(index=colunas, data=arvore_decisao.feature_importances_)
colunas_importantes = colunas_importantes * 100
display(colunas_importantes)
```
